```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests

plt.style.use('ggplot')

requests

df = pd.DataFrame(requests.get('https://vigilo.bapav.org/get_issues.php?').json())
```

```python
# Récupération des catégories
df_categories = pd.read_csv('../data_collection/categories.csv')

# Enregistrements de la catégorie pour chaque signalement
incidents = df['categorie'].tolist()


# Créer un histogramme à partir des données et chaque bar doit être assigner à un seul nombre
plt.hist(incidents, align='left', rwidth=0.5) 

# Ajouter des labels et un titre
plt.xlabel('Numéro de catégorie')
plt.ylabel('Nombre de signalements')
plt.title('Nombre de signalements par catégorie')

# Afficher l'histogramme
plt.show()

# Affichage des categories
df_categories.head(10)
```
